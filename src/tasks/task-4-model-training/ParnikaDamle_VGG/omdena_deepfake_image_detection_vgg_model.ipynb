```python
IMAGE_SIZE = [224,224] #since all of our images have been resized to [160,160]
```

```python
vgg = VGG16(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
for layer in vgg.layers:
  layer.trainable = False
```

```python
keras_merged_train_folders = os.listdir(base_train_path+'\\selected_train')
```

```python
print(keras_merged_train_folders)
```

```python
x = Flatten()(vgg.output)
prediction = Dense(2, activation= 'softmax')(x)
model = Model(inputs=vgg.input,outputs=prediction)
model.summary()
```

```python
#shutil.rmtree(base_train_path+'\\selected_train\\'+'.ipynb_checkpoints')
```

```python
train_datagen = ImageDataGenerator(
    preprocessing_function= preprocess_input,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
)
```

```python
test_datagen = ImageDataGenerator(
    preprocessing_function= preprocess_input,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
)
```

```python
val_datagen = ImageDataGenerator(
    preprocessing_function= preprocess_input,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    horizontal_flip=True,
)
```

```python

train_set = train_datagen.flow_from_directory(base_train_path+'\\selected_train',
                                              target_size = (224,224),
                                              batch_size = 64,
                                              class_mode = "categorical")
```

```python
test_set = test_datagen.flow_from_directory(base_test_path+'\\selected_test',
                                              target_size = (224,224),
                                              batch_size = 64,
                                              class_mode = "categorical")
```

```python
val_set = val_datagen.flow_from_directory(base_val_path+'\\selected_val',
                                              target_size = (224,224),
                                              batch_size = 64,
                                              class_mode = "categorical")
```

```python
from keras import optimizers

model.compile(optimizer= 'adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
from datetime import datetime
from keras.callbacks import ModelCheckpoint

checkpoint = ModelCheckpoint(filepath=base_train_path+'\\selected_train\\'+'best_model.h5', verbose=2, save_best_only=True)

callbacks =[checkpoint]

start = datetime.now()

model_history = model.fit(
    train_set,
    validation_data= val_set,
    epochs=5,
    validation_steps=32,
    callbacks=callbacks, verbose=2)

duration = datetime.now() - start
print("training completed in time: ", duration)
```

```python
batch_size = 32
result = model.evaluate(test_set,batch_size=batch_size)
print("test_loss, test accuracy",result)
```
