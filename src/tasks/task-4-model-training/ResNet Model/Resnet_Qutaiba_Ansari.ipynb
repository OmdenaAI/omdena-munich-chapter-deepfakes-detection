```python
from tensorflow.keras.layers import LeakyReLU, Dropout, Flatten, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
from tensorflow.keras.applications import ResNet50

# Define the input shape
input_shape = (160, 160, 3)

# Load the pre-trained model
base_model = ResNet50(input_shape=input_shape, weights="imagenet", include_top=False)

# Add custom layers on top of the pre-trained model
x = base_model.output
x = Flatten()(x)
x = Dense(1024, kernel_regularizer=l2(0.01))(x)
x = LeakyReLU(alpha=0.1)(x)
x = Dropout(0.5)(x)
x = Dense(512, kernel_regularizer=l2(0.01))(x)
x = LeakyReLU(alpha=0.1)(x)
x = Dropout(0.5)(x)
x = Dense(1, activation="sigmoid")(x)

# Define the model
model = Model(inputs=base_model.input, outputs=x)

# Freeze the layers of the pre-trained model
for layer in base_model.layers:
  layer.trainable = False
```

```python
model.compile(tf.keras.optimizers.Adam(learning_rate=0.001), loss="binary_crossentropy", metrics=["accuracy"])
steps_per_epoch= 47995//32
validation_steps= 6000//16
# Train the model
#model.fit(train_data, epochs=10)
history = model.fit(
train_data,
callbacks=callbacks_list,
steps_per_epoch=steps_per_epoch,
epochs=20,
validation_data=validation_data,
validation_steps=validation_steps)
```
